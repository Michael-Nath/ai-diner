```python
import itertools
Q_values = dict()
num_dining_days = 7
num_diners = 10
num_dining_halls = 5

combinations = itertools.product(range(num_dining_halls), repeat=num_diners)
```

```python
for s in range(num_dining_days):
    print(s)
    combinations = itertools.product(range(num_dining_halls), repeat=num_diners)
    for p in combinations:
        a  = ''.join(str(i) for i in p)
        Q_values[(s, a)] = 0 
```

```python
print(Q_values[(1, '0000')])
```

```python
def get_max_Q_value(Q_values, day):
    a = ""
    best_q_value = -np.inf
    for pair in Q_values.keys():
        if pair[0] != day: continue
        if Q_values[pair] > best_q_value:
            best_q_value = Q_values[pair]
            a = pair[1]
    a = [int(num) for num in a]
    return best_q_value, a
```

```python

env =  DiningHallEnv(num_diners=num_diners, num_dining_days=7, num_dining_halls=num_dining_halls, num_dining_times=10)
obs = env.reset()

for _ in range(100_000):
    cur_day = obs["day"]
    if np.random.binomial(1, 0.45):
        a = np.random.randint(0, num_dining_halls, num_diners)
    else:
        _, a = get_max_Q_value(Q_values, cur_day)
    obs, reward, done = env.step(a)
    max_q, _ = get_max_Q_value(Q_values, obs["day"])
    Q_values[(cur_day, "".join([str(i) for i in a]))] += 0.05 * (reward + 0.95 * max_q - Q_values[(cur_day, "".join([str(i) for i in a]))])
        
```
